```python
import math as mth
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import scipy as sci
from scipy.optimize import bisect
from scipy.integrate import quad

#Взят вариант 22 вместо варианта 2 для перерасчета для Si
materialName = "Si"
Tmax = 700
Tmin = 70
Nd1 = 0
Nd2 = 0 
Ed1 = 0
Ed2 = 0

Na1 = 1e13
Na2 = 1e14
Ea1 = 0.05
Ea2 = 0.1

Nt = 1e14
σn = 1e-16
σp = 1e-14
Et = 0.8
S = 3.2e4
h = 0.06
a = 5
l = 5
λ = 1.3e-4
Ps = 0.2
R = 0.36
α = 1e4
Fm = 4.5

Eg0 = 1.156
bt = 1108
at  = 7.021e-4
eps = 11.8
Cl = 1.927e12
Eact = 10.20
k = 8.62e-5
m0 = 9.11e-31
mp = 0.56
mn = 1.08
Xsame = 4.05
eps0 = 8.85e-14
workMe = 4.5 
tox = 100e-7
epsOx = 3.725

bb = 2.13
aa  = 9.6
cc = 12/5
q = -1.6e-19

VBO = 4.4
CBO = 3.5  

xTempRange = np.arange(Tmin, Tmax+1, 1)
nRange = np.arange(-10, 10, (20/(xTempRange.size)))

tempComnind = 230

plt.rcParams['figure.dpi'] = 100
plt.rcParams["figure.figsize"] = (9.6,5.4) 
plt.rcParams["axes.grid"] = True
```

```python
#расчет Eg 
Eg = Eg0 - (at*(xTempRange**2))/(bt+xTempRange)
plt.xlabel("Температура Т, К")
plt.ylabel("Энергия, эВ")
plt.title('Ширина запрещённой зоны')
plt.plot(xTempRange, Eg , label = 'Eg') 

plt.legend();
```

```python
#расчет Nc 
Nc = 2.5e19 * mn**(3/2) * (xTempRange/300)**(3/2)
#расчет Nv 
Nv = 2.5e19 * mp**(3/2) * (xTempRange/300)**(3/2)
plt.xlabel("Температура Т, К")
plt.ylabel("Плотность квантовых состояний см^-3")
plt.title('Плотность квантовых состояний')
plt.plot(xTempRange, Nc , label = 'Nc') 
plt.plot(xTempRange, Nv , label = 'Nv') 
plt.semilogy()
plt.legend(); 
```

```python
#расчет интеграла Ферми порядка 1/2 
fermi = ((np.sqrt(np.pi/2))**(-1))*((((1.5*(2**1.5))/(bb+nRange+(((np.fabs(nRange-bb))**(1/cc))+aa)**1.5))+((np.exp(-nRange))/((np.sqrt(np.pi))/(2))))**(-1))
plt.xlabel("Тетта")
plt.ylabel("1/2 F")
plt.title('График проверки интеграла Ферми порядка 1/2')
plt.plot(nRange, fermi) 
plt.semilogy();
```

```python
# Зонная диаграмма
# Дно зоны проводимости
Ec = Eg.copy()
# Потолок валентной зоны
Ev = []
for i in range(xTempRange.size):
    Ev.append(0)

def Ferm(nn):
    fremi = (((np.sqrt(np.pi/2))**(-1))*((((1.5*(2**1.5))/(bb+nn+(((np.fabs(nn-bb))**(1/cc))+aa)**1.5))+((np.exp(-nn))/((np.sqrt(np.pi))/(2))))**(-1)))
    return fremi
Ei = Eg/(k*xTempRange)

def f(nn,i):
    p = Nv[i]*Ferm(-nn - Ei[i])
    n = Nc[i]*Ferm(nn)
    na1 = Na1 / (1 + 2 * np.exp(-nn + (Ea1-Eg[i])/(k*xTempRange[i])))
    na2 = Na2 / (1 + 2 * np.exp(-nn + (Ea2-Eg[i])/(k*xTempRange[i])))
    return p-n-na1-na2

Fimax = Ec[630]/2 + 0.75*k*xTempRange[630]*mth.log((mp/mn),(mth.exp(1)))
nn_start = (Fimax - Eg[xTempRange.size-1]) / (k*xTempRange[xTempRange.size-1])
ETA = [1]
ETA[0] = sci.optimize.fsolve(f, nn_start , args = (630),xtol=1e-11)[0]
for i in range(1,xTempRange.size):
    ETA.append(sci.optimize.fsolve(f, ETA[i-1] , args = (630-i),xtol=1e-11)[0])
```

```python
#Собственный уровенль Ферми
Es = Eg/2 + 0.75*k*xTempRange*mth.log((mp/mn),(mth.exp(1)))
#Уровень ферми
F = k*ETAn*xTempRange+Eg
ea1=[]
ea2=[]
for i in range(xTempRange.size):
    ea1.append(Ea1)
    ea2.append(Ea2)
```

```python
plt.xlabel("Т,K")
plt.ylabel("E, эВ")
plt.title('Зонная диаграмма')
plt.plot(xTempRange, Ev, label = 'Ev')
plt.plot(xTempRange, Ec, label = 'Ec')
plt.plot(xTempRange, Es, label = 'Ei') 
plt.plot(xTempRange, ea1, label = 'Ea1')
plt.plot(xTempRange, ea2, label = 'Ea2')
plt.plot(xTempRange, F, label = 'F')
    
plt.legend();
```

```python
#Расчет концентраций электронов, дырок и собственной концентрации
p = Nv*Ferm(-ETAn-(Eg/(k*xTempRange)))
n = Nc*Ferm(ETAn)
ni = np.sqrt(n*p)

plt.xlabel("Т,K")
plt.ylabel("концентрация, см^-3")
plt.title('Концентрация')
plt.plot(1/xTempRange, p, label = 'p')
plt.plot(1/xTempRange, n, label = 'n')
plt.plot(1/xTempRange, ni, label = 'ni')

plt.semilogy()
plt.legend();
```

```python
#Расчет концентраций ионов акцепторов
na1 = Na1 / (1 + 2 * np.exp(-ETAn + (Ea1-Eg)/(k*xTempRange)))
na2 = Na2 / (1 + 2 * np.exp(-ETAn + (Ea2-Eg)/(k*xTempRange)))

plt.xlabel("Т,K")
plt.ylabel("концентрация, см^-3")
plt.title('Концентрации ионов акцепторов')
plt.plot(1/xTempRange, na1, label = 'na1')
plt.plot(1/xTempRange, na2, label = 'na2')

plt.semilogy()
plt.legend();
```

```python
ff = np.fabs(p-n-na1-na2)
ff[ff == 0] = 1e-5

plt.xlabel("Т,K")
plt.ylabel("численная ошибка, см^-3")
plt.title('электронейтральность')
plt.plot(xTempRange, ff, label = 'ff')

plt.semilogy();
```

```python
#Расчет положения уровня Ферми от зарядового уравнения
FermiLevelQeq = k * xTempRange[230] * np.log(Nv[230]/(na1[230]+na2[230])) + Ev[230]
#Расчет положения уровня Ферми от потенциала Ферми
potFermi = k*xTempRange[230]*np.log(((na1[230]+na2[230])/ni[230]))
FermiLevelPot = Es[230] - qpos*potFermi
FCounted = F[230]
print(FCounted," рассчитанынй уровень Ферми при 300K")
print(FermiLevelPot," уровень Ферми от Потенциала Ферми при 300K")
print(FermiLevelQeq," уровень Ферми от основных носителей зарядов при 300K")
#Работа выхода полупроводника
WorkOut = Xsame + Eg[230] - F[230]
print(WorkOut," работа выхода полупроводника при 300K")
#Расчет напряжения плоских зон
Vfb = workMe - WorkOut
print(Vfb," напряжение плоских зон полупроводника при 300K")
```

```python
#Расчет зависимости удельного поверхностного заряда от поверхностного потенциала
def fsFunc(fis):
    return (0.026*p[230]*(np.exp(-(fis / 0.026)) - 1)) + (0.026*n[230]*(np.exp(fis/0.026) - 1)) + (fis * (na1[230]+na2[230]))
def QsFunc(fis):
    return (-np.sign(fis) * ((2 * qpos * eps*eps0* fsFunc(fis)) ** (1 / 2)))
fisArray = np.arange(Ev[230]-F[230], Ec[230]-F[230], 0.001761)
fs = fsFunc(fisArray)
Qs = QsFunc(fisArray)

plt.xlabel("Ψs, эВ")
plt.ylabel("Qs, Кл/см^2")
plt.title('Поверхностный заряд')
plt.plot(fisArray, Qs);
```

```python
#Расчет зависимости поверхностного потенциала ψs от напряжения затвор—подложка Vgb, используя численные методы.
Coxp=epsOx*eps0/tox
def SPE(fisNew,Vgb):
    return Vfb+fisNew-QsFunc(fisNew)/Coxp-Vgb
Vgb=np.linspace(SPE(Ev[230]-F[230],0),SPE(Ec[230]-F[230],0),631)


fisss=np.array([])
anew=Ev[230]-F[230]
bnew=Ec[230]-F[230]
for val in Vgb:
    fisss = np.hstack((fisss, bisect(f=SPE,a=anew,b=bnew, args=val, xtol=1e-14)))

plt.xlabel("Vgb, В")
plt.ylabel("Ψs, эВ")
plt.title('Поверхностный потенциал')
plt.plot(Vgb,fisss, label = 'Vgb');
plt.axvline(x=Vfb,color='red', label = 'Vfb')
plt.axvline(x=SPE(Ev[230]-F[230],0),color='green', label = 'SPE(Ev-F)')
plt.axvline(x=SPE(Ec[230]-F[230],0),color='yellow', label = 'SPE(Ec-F)')
plt.legend();
```

```python
Qnew = (fisss+Vfb-Vgb)*Coxp

plt.xlabel("Vgb, В")
plt.ylabel("Qs, Кл/см^2")
plt.title('Поверхностный заряд')
plt.plot(Vgb,Qnew, label = 'Qs')
plt.axvline(x=Vfb,color='red', label = 'Vfb')
plt.axvline(x=SPE(Ev[230]-F[230],0),color='green', label = 'SPE(Ev-F)')
plt.axvline(x=SPE(Ec[230]-F[230],0),color='yellow', label = 'SPE(Ec-F)')
plt.legend();
```

```python
#Пункт 1
nfis = lambda fisss: n[230] * np.exp(fisss / 0.026)
pfis = lambda fisss: p[230] * np.exp(-fisss / 0.026)
Efis = lambda fisss: np.sign(fisss) * np.sqrt(2 * (qpos) * fsFunc(fisss)/ (eps*eps0))

Ifis = lambda fisss: 1-np.abs(np.sign(fisss))
Sp= lambda f,ep: (f+np.sqrt(f**2+4*ep**2))/2
Sn= lambda f,ep: (f-np.sqrt(f**2+4*ep**2))/2

integr_inv =  lambda fisss: (qpos * n[230] * (np.exp(fisss/0.026) - 1))/((np.sign(fisss) * np.sqrt(2 * (qpos) * fsFunc(fisss)/ (eps*eps0)))+Ifis(fisss))
integr_bulk =  lambda fisss: (-qpos * p[230] * (np.exp(-fisss/0.026) - 1))/((np.sign(fisss) * np.sqrt(2 * (qpos) * fsFunc(fisss)/ (eps*eps0)))+Ifis(fisss))
```

```python
Qi = np.array([])
Qi = np.zeros(631) 
Qb = np.array([])
Qb = np.zeros(631)

for i in range(631):
        result_i, error = quad(integr_inv, fisss[i], 0)
        Qi[i]= Sn(result_i, 1e-14)

for i in range(631):
        # result_i, error = quad(integr_bulk, fisss[i], 0)
        # Qb[i]= result_i
        result_i, error = quad(integr_bulk, fisss[i], 0)
        Qb[i]= Sp(result_i, 1e-14)
```

```python
#из методических указаний 1 к КМ-4 - выбор координат
integrY = lambda fisss: 1 / Efis(fisss)

def YCoordCompute(fisss):
    if fisss == 0:
        y = np.linspace(0, 150, 101) * 1e-7 
        fis = 0 * y
        return y, fisss 
    fis1 = np.linspace(fisss, fisss * 0.5, 21) 
    fis2 = np.logspace(np.log10(np.abs(fisss * 0.5)), np.log10(np.abs(fisss * 1e-3)), 101)
    if fisss < 0:
        fis2 = -1 * fis2
    fis = np.hstack((fis1, fis2[1:]))
    y = np.array([])
    for value in fis:
        YTemporal, error = quad(integrY, value, fisss)
        y = np.hstack((y, YTemporal))
    return y, fis
```

```python
newFis = (Ev[230] - F[230]) 
Vgb_new=SPE(newFis,0)
FisOx = Vgb_new - newFis - Vfb
y, newFisExtra = YCoordCompute(newFis)
y = y / 100 * 1e9
toxnm = tox / 100 * 1e9

nNew = n[230] * np.exp(newFisExtra / 0.026)
pNnew = p[230] * np.exp(-newFisExtra / 0.026)

QEqZero = pNnew - nNew - Na1 - Na2

# для зонной диаграммы 
fig, ax = plt.subplots() 
plt.figure(1)
ax.set_title('Зонная диаграмма(обогащение)')
plt.xlabel('x (нм)', fontsize=14)
plt.ylabel('Энергия , эВ', fontsize=14)
plt.grid(True)

# зонная диаграмма п\п
plt.plot(y, Ev[230] - newFisExtra, 'b', label = "Ev")
plt.plot(y, Es[230] - newFisExtra, 'g--', label = "Ei")
plt.plot(y, Ec[230] - newFisExtra, 'orange', label = "Ec")

# уровень ферми
plt.plot(y, 0 * y + F[230], 'k',  label = "F")
plt.legend(loc='best', prop={'size': 16})

#диаграмма оксида
plt.plot([0, 0], [Ev[230] - newFis - VBO, Ec[230] - newFis + CBO], 'r')
plt.plot([-toxnm, -toxnm],[Ev[230] - newFis - VBO - FisOx, Ec[230] - newFis + CBO - FisOx],  'r')
plt.plot([-toxnm, 0], [Ev[230] - newFis - VBO - FisOx, Ev[230] - newFis - VBO], 'r')
plt.plot([-toxnm, 0], [Ec[230] - newFis + CBO - FisOx, Ec[230] - newFis + CBO], 'r')
plt.legend(loc='best', prop={'size': 16})

# уровень ферми метала
plt.plot([-toxnm - 100, -toxnm], [F[230] - Vfb - newFis - FisOx, F[230] - Vfb - newFis - FisOx], 'k')

fig, ax = plt.subplots() 
plt.figure(2)
plt.xlabel('x (нм)', fontsize=14)
plt.ylabel('(1 / см$^3$)', fontsize=14)
plt.grid(True)
ax.set_title('Концентрации(обогащение)')
plt.semilogy(y, nNew, label='n', linewidth=2)
plt.semilogy(y, pNnew, label='p', linewidth=2)
plt.semilogy(y, Na1+Na2 + 0 * y, label='Na')
plt.legend(loc='best', prop={'size': 16})

#обемный заряд
fig, ax = plt.subplots() 
ax.set_title('Объемный заряд(обогащение)')
plt.figure(3)
plt.xlabel('x (нм)', fontsize=14)
plt.ylabel('$|\\rho/q|$ (1 / см$^3$)', fontsize=16)
plt.grid(True)
tmp = plt.semilogy(y, np.abs(QEqZero), label='$\\rho/q$')

plt.legend();
```
